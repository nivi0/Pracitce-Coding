```python
#12) Get Node Value

def getNode(head, positionFromTail):
    
    pt1 = head
    pt2 = head
    
    for i in range(positionFromTail):
        pt1 = pt1.next
        
    while pt1.next != None:
        pt1 = pt1.next
        pt2 = pt2.next
        
        
    return pt2.data
```

```python
#13) Delete duplicate-value nodes from a sorted linked list

def removeDuplicates(head):
    if head == None :
        return None
    
    prev = head
    cur = head.next
    while cur != None:
        if cur.data == prev.data:
            prev.next = cur.next
            cur = prev.next
        else:
            prev = prev.next
            cur = cur.next

    return head
```

```python
#17) Reverse a doubly linked list
def reverse(head):

    while head.next != None :
        head.next, head.prev, head = head.prev, head.next, head.next
    
    head.next, head.prev = head.prev, None
    
    return head
```

```python
class Node:
    def __init__(self, info): 
        self.info = info  
        self.left = None  
        self.right = None 
        self.level = None 

    def __str__(self):
        return str(self.info) 

class BinarySearchTree:
    def __init__(self): 
        self.root = None

    def create(self, val):  
        if self.root == None:
            self.root = Node(val)
        else:
            current = self.root
         
            while True:
                if val < current.info:
                    if current.left:
                        current = current.left
                    else:
                        current.left = Node(val)
                        break
                elif val > current.info:
                    if current.right:
                        current = current.right
                    else:
                        current.right = Node(val)
                        break
                else:
                    break

"""
Node is defined as
self.left (the left child of the node)
self.right (the right child of the node)
self.info (the value of the node)
"""
```

```python
#30) BST : Lowest Common Ancestor
def lca(root, v1, v2):
  #Enter your code here
  if v1 > v2:
    v1, v2 = v2, v1  #this is to make sure that v1 < v2
    
  while True:
    if v1 < root.info and v2 < root.info:
        root = root.left
    elif v1 > root.info and v2 > root.info :
        root = root.right
    else :
        return root
```

```python
#33) Equal Stack
def equalStacks(h1, h2, h3):
    # Write your code here
    
    #reverse all the stacks
    h1 = h1[::-1]
    h2 = h2[::-1]
    h3 = h3[::-1]
    #sum of heights of cylinder in each stack
    sum1 = sum(h1)
    sum2 = sum(h2)
    sum3 = sum(h3)
    
    while True:
        minheight = min(sum1, sum2, sum3)
        
        #base case
        if minheight == 0:
            return 0
        
        #general case
        if minheight < sum1 :
            sum1 -= h1.pop()
        if minheight < sum2 :
            sum2 -= h2.pop()
        if minheight < sum3 :
            sum3 -= h3.pop()
            
        if sum1 == sum2 == sum3 :
            return sum1
```
